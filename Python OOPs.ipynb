17.What is the difference between a class variable and an instance variable?
Ans.The main difference between a class variable and an instance variable is that class variables are shared by all instances of a class, while instance variables are unique to each instance:
Class variables
A single copy of a class variable exists, regardless of how many instances of the class exist. Class variables are useful for sharing data between all instances of a class, such as keeping track of the number of instances created or storing a constant value.
Instance variables
Instance variables are created when an object is instantiated and are accessible to all the constructors, methods, or blocks in the class. Instance variables are useful for storing data that is specific to each instance, such as the name or age of a person.

18.What is multiple inheritance in Python?
Ans.Multiple inheritance in Python is a feature that allows a class to inherit from more than one parent class. This means that the child class can inherit attributes and methods from multiple sources, combining their functionalities.

Key points:
Syntax:
To use multiple inheritance, simply list the parent classes in the class definition, separated by commas (e.g., class Child(Parent1, Parent2):).
Method Resolution Order (MRO):
Python uses a specific algorithm, called the C3 linearization algorithm, to determine the order in which methods are resolved in the case of inheritance conflicts. This ensures that the inheritance hierarchy is traversed in a predictable manner.
super() function:
The super() function is used to call methods from parent classes, which is especially useful in multiple inheritance scenarios where you need to access functionality from multiple parent classes.
Diamond problem:
Multiple inheritance can lead to the "diamond problem", where a class inherits from two parent classes that share a common ancestor. Python resolves this ambiguity using the MRO.

22.What is the difference between @staticmethod and @classmethod in Python?
ans. In Python, both @staticmethod and @classmethod are decorators used to define methods that operate on the class level rather than on instances of the class. However, they have key differences:
@staticmethod
Purpose: Used for methods that don't need to access or modify the class state or any instance state. They act like regular functions but are logically grouped within the class.
Parameters: Doesn't take any implicit parameters like self or cls.
Access: Can't access class attributes or instance attributes directly.

@classmethod
Purpose: Used for methods that need to access or modify the class state but not any specific instance state. They can be used for factory methods or other operations related to the class itself.
Parameters: Takes cls as the first parameter, which refers to the class itself.
Access: Can access class attributes but not instance attributes.

```python
#10 Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
```
